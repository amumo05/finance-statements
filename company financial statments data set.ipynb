```python
def read(filename: str) -> List[CompanyRecord]:
    """Reads the CSV file and returns a list of CompanyRecord entries."""
    loc: List[CompanyRecord] = []

    with open(filename, newline='', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # skip header

        for row in reader:
            year_s = row[0]
            mcap_s = row[3]
            rev_s = row[4]
            npm_s = row[18]

            if year_s and mcap_s and rev_s and npm_s:
                try:
                    loc.append(
                        CompanyRecord(
                            year=int(year_s),
                            company=row[1],
                            category=row[2],
                            market_cap_b=float(mcap_s),
                            revenue=float(rev_s),
                            net_profit_margin=float(npm_s)
                        )
                    )
                except ValueError:
                    pass  # skip bad rows

    return loc
```

```python
import matplotlib.pyplot as plt

def make_summary(p: Produced) -> str:
  """
  consumes the Produced list and returns the textual summary of that companies data
  which has the information about the Year-over-year NPM changes, 3-year moving averages ,
  CAGR of Market Cap,and the  Correlation (Market Cap vs Revenue) value of the given Produced data  
  """
  #return "" #stub 
  #template is based on Produced
  yoy_text = ""
  i = 0
  while i < len(p.yoy):
    if i != 0:
        yoy_text += ", "
    yoy_text += str(round(p.yoy[i], 2))
    i += 1

     # 3-year moving averages
    ma3_text = ""
    for i in range(len(p.ma3)):
       if i != 0:
          ma3_text += ", "
       ma3_text += str(round(p.ma3[i], 2))

    # Round CAGR and correlation
    cagr_rounded = round(p.cagr, 2)
    corr_rounded = round(p.corr, 2)

    # Build summary string
    text = "Year-over-year NPM changes (%): " + yoy_text + ". "
    text += "3-year moving averages: " + ma3_text + ". "
    text += "CAGR of Market Cap: " + str(cagr_rounded) + "%. "
    text += "Correlation (Market Cap vs Revenue): " + str(corr_rounded)

    return text

def plot_company(p: Produced, company: str) -> None:
    """
    Analyzes the dataset and produces a net profit margin plot
    for the selected company.
    """
    #return None #stub 
    #template is based on visualization 
    plt.figure(figsize=(12, 6))
    
    plt.plot(p.years, p.npm, marker='o', label="Net Profit Margin (%)",color="#2ca02c") #this will plot the net-per-margin line 
                                                                                      # "#D3ECCC"<- pastel green
    if p.ma3:  # Plots 3-year moving average
        # The moving average corresponds to years starting from index 2
        plt.plot(p.years[2:], p.ma3, marker='x', linestyle='--', label="3-Year Moving Average",color="#ff69b4") # "#FFD9E0"<- pastel pink 
                                                                                            
    plt.xlabel("Year")
    plt.ylabel("Net Profit Margin (%)")
    plt.title(company + " Profitability Trend")
    plt.legend()
    plt.xticks(p.years)
    plt.show()


```

##### <b>MAIN FUNCTIONS</b>

```python
def main(filename: str, company: str) -> None:
    """
    Reads the financial CSV file,
    analyzes the chosen company,
    plots the line chart with NPM and 3-  moving average,
    and returns the textual summary as a string.
    """
    # Template from HtDAP, based on composition
    #runs all the code here 
    records = read(filename)                 # read ONCE
    analysis = analyze(records, company)     # analyze ONCE

    print(make_summary(analysis))
    plot_company(analysis, company)

    
    #calculates the given  companies information out of the data 
def analyze(loc:CompanyRecordList, company: str) -> Produced:
    """
    Selects only the chosen company's rows,
    sorts them by year,
    computes NPM series, YoY changes,
    moving average, CAGR, and correlation.
    Returns a Produced.
    """
    #return [] #stub 
    #template is based on CompanyRecordList and an additional parameter company
    #filtered stores the produced List
    
    filtered = []
    for cr in loc:
        if cr.company == company:
            filtered = filtered + [cr]

    #helper functions 
    sorted_list = sort_by_year(filtered)
    years = extract_years(sorted_list)
    npm = extract_npm(sorted_list)
    yoy = compute_yoy(npm)
    ma3 = moving_average_3(npm)
    cagr = compute_cagr(sorted_list)
    corr = pearson_corr(sorted_list)

    return Produced(years, npm, yoy, ma3, cagr, corr)


# HELPER FUNCTIONS 

def sort_by_year(loc: CompanyRecordList) -> CompanyRecordList:
    """
    consumes CompanyRecordList and returns a new list sorted by year 
    """
    #return [] #stub 
    #template is based on CompanyRecordList 
    #lst stores the sorted list by year for the given CompanyRecordList
    lst = loc[:]   #type:CompanyRecord
    n = len(lst)
    for i in range(n):
        for j in range(n - 1):
            if lst[j].year > lst[j+1].year:
                temp = lst[j]
                lst[j] = lst[j+1]
                lst[j+1] = temp
    return lst

def extract_years(loc: CompanyRecordList) -> YearList:
    """
    consumes CompanyRecordList and returns a YearList
    """
    #return [] #stub 
    #template is based on CompanyRecordList
    #result stores the list of years 
    result = [] #type:int
    for cr in loc:
        result.append(cr.year)
    return result


def extract_npm(loc: CompanyRecordList) -> FloatList:
    """
    consumes a CompanyRecordList and returns a new list consisting the net per margin value 
    """
    #return [] #stub 
    #template is based on CompanyRecordList 
    #result stores the list of npm values 
    result = [] #type:float
    for cr in loc:
        result.append(cr.net_profit_margin)
    return result


def compute_yoy(nums: FloatList) -> FloatList:
    """
    consumes the FloatList and returns a new list consisting the revenue values 
    """
    #return [] #stub 
    #template is based on FloatList 
    #result stores the list of yoy values 
    result = [] #type:float
    
    i = 1  # starts the test from the second element
    while i < len(nums):
        prev = nums[i-1]
        curr = nums[i]
        if prev == 0:
            result.append(0.0)
        else:
            result.append(((curr - prev) / prev) * 100)
        i += 1
    return result


def moving_average_3(nums: FloatList) ->FloatList:
    """
    consumes the FloatList and returns a new list consisting the revenue values 
    """
    #return [] #stub 
    #template is based on FloatList 
    #result stores the list of moving average over 3 years values 
    
    result = [] #type:float
    
    i = 0 
    while i + 2 < len(nums):
        sum3 = nums[i] + nums[i+1] + nums[i+2]
        result.append(sum3 / 3)
        i += 1
    return result


def compute_cagr(loc: CompanyRecordList) -> float:
    """
    consumes the CompanyRecordList and returns the compound annual growth rate 
    """
    #return 0.0 #stub 
    #template is based on CompanyRecordList
    #n stores the calculated value 
    if len(loc) < 2:
        return 0.0
    start = loc[0].market_cap_b
    end = loc[-1].market_cap_b
    n = 0  #type:float
    temp = loc
    for _ in temp:
        n += 1
    n = n  # number of years
    return ((end / start) ** (1 / (n - 1))) - 1


def pearson_corr(loc: CompanyRecordList) -> float:
    """
    consumes CompanyRecordList and returns the Pearson correlation coefficient of that particular record list 
    """
    #return 0.0 #stub 
    #template is based on CompanyRecordList
    #n stores the  Pearson correlation coefficient 
    
    n = 0 #type:float
    
    for _ in loc: #the for-statment is going to loop through every item in loc 
        n += 1
    if n == 0:
        return 0.0

    sum_x = 0.0
    sum_y =0.0
    
    for cr in loc:
        sum_x =sum_x+ cr.market_cap_b
        sum_y =sum_y+ cr.revenue

    mean_x = sum_x / n
    mean_y = sum_y / n

    num = 0.0
    denom_x = 0.0
    denom_y = 0.0
    for cr in loc:
        dx = cr.market_cap_b - mean_x
        dy = cr.revenue - mean_y
        num =num + dx * dy
        denom_x =denom_x + dx * dx
        denom_y =denom_y + dy * dy

    if denom_x == 0 or denom_y == 0:
        return 0.0
    return num / ((denom_x * denom_y) ** 0.5)
```
